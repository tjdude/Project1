```python
from time import sleep
import random
from selenium.webdriver.common.by import By

xp = '/html'
news_box = driver.find_element(By.XPATH, xp)  # 드래그가 있는 요소

#스크롤 함수
def scrolldown(max_scroll=210):
    prev_height = -1
    for _ in range(max_scroll):
        current_height = driver.execute_script("return document.body.scrollHeight")
        
        if current_height == prev_height:  # 더 이상 스크롤되지 않으면 종료
            print("✅ 스크롤 완료! 더 이상 새로운 기사가 없습니다.")
            return False
        
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        prev_height = current_height
        sleep(random.randint(2, 3))  # 2~3초 쉬었다가 반복
    
    return True

#저장 리스트 초기화
titles = []
contents = []
max_articles = 2000  # 크롤링할 기사 개수
prev_article_count = 0  # 스크롤 전 기사 개수 저장

while len(titles) < max_articles:
    if not scrolldown():  # 스크롤 끝까지 내리면 자동 종료
        break
        
    # 뉴스 제목과 요약을 한 번에 가져오기
    title_elements = driver.find_elements(By.CSS_SELECTOR, "a.news_tit")
    content_elements = driver.find_elements(By.CSS_SELECTOR, "a.api_txt_lines.dsc_txt_wrap")

    print(f"✅ 현재 뉴스 개수: {len(title_elements)}개")

    
    # ✅ 스크롤 후 새로운 기사가 없으면 종료
    if len(title_elements) == prev_article_count:
        print("✅ 더 이상 새로운 기사가 없습니다. 크롤링 종료!")
        break

    
    prev_article_count = len(title_elements)  # 현재 기사 개수 업데이트

    for i in range(min(len(title_elements), len(content_elements))):
        try:
            title = title_elements[i].get_attribute('innerText')
            content = content_elements[i].get_attribute('innerText')

            titles.append(title)
            contents.append(content)  # 리스트에 같은 순서로 저장

            # 목표 개수 도달 시 종료
            if len(titles) >= max_articles:
                break

        except Exception as e:
            print(f"🚨 {i}번째 뉴스 `stale element` 발생, 재시도...")
            sleep(2)  # 2초 대기 후 재시도
            title_elements = driver.find_elements(By.CSS_SELECTOR, "a.news_tit")
            content_elements = driver.find_elements(By.CSS_SELECTOR, "a.api_txt_lines.dsc_txt_wrap")
            continue  # 다음 뉴스로 이동

# Pandas DataFrame 변환
df = pd.DataFrame({"제목": titles, "내용": contents})
```

```python
df = pd.DataFrame({"제목": titles, "내용": contents})

# CSV 저장
df.to_csv("naver_news(4).csv", index=False, encoding="utf-8-sig")
print("CSV 파일 저장 완료: naver_news(4).csv")

# 저장된 데이터 확인
df_check = pd.read_csv("naver_news(4).csv")
print("✅ 저장된 데이터 개수:", len(df_check))
print(df_check.head())  # 상위 5개 출력
```
