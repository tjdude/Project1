```python
#형태소 분석 및 불용어 제거 - 토큰 텍스트 clean 컬럼에 할당
import pandas as pd
from mecab import MeCab
import re

mecab = MeCab()

use_tags = ['NNP', 'NNG', 'NP', 'IC', 'MAG','SL','VA','VV']
stopwords = stopwords = ['',"은", "는", "이", "가", "을", "를", "의", "도", "에", "에서", "그리고", "그러나", "하지만",
             "따라서", "또한", "즉", "때문에", "이외에도", "특히", "해", "지난해", "시", "뉴스", "연합뉴스", "KBS", "MBC", "SBS", "JTBC", "조선일보",
             "기사", "보도", "속보", "기자", "취재", "제공", "환자", "건강", "질병", "치료", "예방", "의사",
             "연구", "발표", "결과", "고혈압", "혈압", "저혈압", "심장", "뇌졸중", "영향", "관리",
             "최고", "최저", "수준", "2배", "배", "퍼센트", "차이", "수축기", "확장기", "보건복지부", "보험",
             "사용", "주요", "의학", "최근", "국제", "저널", "시간", "따르", "늘어나", "가치", "부르",
             "초래", "맛있", "현지", "이것", "관련", "질환", "더욱", "부담", "어려움", "시민", "주의",
             "최소", "등록", "교육", "센터", "추진", "프로그램", "전체", "여기", "해당", "가장", "한양대",
             "교수", "중반", "결심", "단순", "필수", "보통", "요인", "단계", "모두", "어머니", "정주영",
            '하남시', '사업가', '증가세', '지참', '만약', '이상무', '경우', '한과', '진료', '의료', '위험',
            '국내', '보고', '대상', '병원장', '지표', '보다', '만자', '유독', '보건소', '진천군',
            '한양', '경쟁', '고창군','증상', '사업', '증가', '발생', '진단', '학회', '병원','효과', '개선', '진행', '사람'
            ,'필요', '운영', '보장', '이번', '삼성', '이상','함께','대표', '개발','중요', '지원', '섭취', '검진','원인',
            '조사', '교실', '생활']


def cleaning(review):

    clean_words = mecab.pos(review)

    clean_words = [w[0]for w in clean_words if w[1] in use_tags]

    clean_words = [word for word in clean_words if word not in stopwords and len(word) > 1]
    clean = ' '.join(clean_words)

    return clean


df['clean'] = df['text'].apply(cleaning) 
df.to_csv('pre_news.csv',index = False)
df
```

```python
#4. clean 텍스트를 빈도로 변환하여 출력
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('pre_news.csv').dropna()

df['clean'] = df['clean'].str.replace("당뇨병", "당뇨")

clean_vec = CountVectorizer(max_df=0.9, min_df=10)

clean_vec.fit(df['clean'])

names = clean_vec.vocabulary_
print(len(names))
print(names)
```

```python
#빈도 그래프
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

cvec = CountVectorizer(max_df=0.9, min_df=5, analyzer = 'word',
                      )
X = cvec.fit_transform(df['clean'])
x = X.toarray()
x.shape
```

```python
names = cvec.get_feature_names_out()
names
```

```python
xsum = np.sum(x, axis=0)
xsum
```

```python
rank_ind = np.argsort(xsum)
rank_ind
```

```python
#상위 20개 키워드
most_n = []
most_v = []

all_keywords = names  # CountVectorizer로 추출된 전체 단어 리스트

for ind in rank_ind[-20: ]:
    most_n.append(names[ind])
    most_v.append(xsum[ind])
print(most_n)
print(most_v)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(6,5))
fcolor = '#212121'
gcolor = '#FF9800'


ax = plt.subplot()
hbars = ax.barh(most_n, most_v, height=0.7, color=gcolor)
ax.set_title('빈도그래프', fontsize=20)
ax.invert_yaxis()
ax.tick_params(axis='both', labelsize=10, labelcolor=fcolor) 

ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.bar_label(hbars, fmt='%g', padding=8, color=gcolor, fontsize=10) 

plt.show()
```

```python
#TF-IDF 그래프
from sklearn.feature_extraction.text import TfidfVectorizer

tvec = TfidfVectorizer(max_df=0.9, min_df=10)
X = tvec.fit_transform(df['clean'])
names = tvec.get_feature_names_out() # 토큰 목록을 배열로 가져옴
x = X.toarray()
#xsum = np.sum(x, axis=0) # 2차원 배열의 벡터를 세로(0번축)으로 합침
xmean = np.mean(x, axis=0)
#sum을 쓸지 mean을 쓸지 잘 선택하기

most_n = []
most_v = []
for ind in np.argsort(xmean)[-15:]: # argsort() 값이 작은 순으로 인덱스가 정렬
    most_n.append(names[ind]) # 토큰과 개수를 순서별로 저장
    most_v.append(xmean[ind])
most_n.reverse()
most_v.reverse()
print(most_n)
print(most_v)
```

```python

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

plt.figure(figsize=(6,5))
fcolor = '#212121'
gcolor = '#FF5722'


ax = plt.subplot()
hbars = ax.barh(most_n, most_v, height=0.7, color=gcolor)
ax.set_title('TF-IDF', fontsize=20)
ax.invert_yaxis() # 축 반대 정렬
ax.tick_params(axis='both', labelsize=10, labelcolor=fcolor)
# 테두리 삭제
for i in 'right left'.split(): 
    ax.spines[i].set_visible(False)

ax.bar_label(hbars, fmt='%.3f', padding=8, color=gcolor, fontsize=10) 

#plt.savefig("TF-IDF.png", dpi=300, bbox_inches="tight")  # 고해상도 저장

plt.show()
```

LDA : 가지고 있는 텍스트를 군집화 시켜서 그 키워드들이 합병증으로 합쳐지는지 시각화로 확인

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

cvec = CountVectorizer(max_df=0.9, min_df=10)
tf = cvec.fit_transform(df['text'])
model = LatentDirichletAllocation(n_components = 4,
                                  max_iter = 10,
                                  learning_method = 'online',
                                  random_state = 0,)

model.fit(tf)
```

```python
import pyLDAvis
import pyLDAvis.lda_model

pyLDAvis.enable_notebook()

pl = pyLDAvis.lda_model.prepare(model, tf, cvec)

pyLDAvis.display(pl)
```
